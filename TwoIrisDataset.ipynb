```python
import numpy as np
from google.colab import files
import matplotlib .pyplot as plt

dataset=np.loadtxt("arquivo.txt", delimiter=" ")
```

```python
entradas=np.delete(dataset, -1, axis=1)
print("")
print(entradas)
xy=np.insert(entradas, 0, values=-1, axis=1)
print(xy)
d=dataset[:, -1]
print(d)
w=np.random.rand(len(xy[0]))
w[0]=1.4

print(f"pesos iniciais = {w}")
```

```python
n=0.1
epoca=0
erro=1

while erro == 1:
  erro=0
  for i in range (0, len(xy)):
    u=sum(xy[i]*w)
    y=1 if u>=0 else 0

    if i >= 50:
      print("virginica")
    else:
      print("versicolor")

    if y != d[i]:
      w=w+(n*(d[i]-y)*xy[i])
      erro=1
    print(25*"=-")
    print(f'potencial de ativação = {u}')
    print(f"saida desejada (d) = {d[i]} saida (y) = {y}")
    print(f"fim da época de de trinamento {epoca+1}")
    print(25*"=-")
    print("")
    if i == 200:
      break

  epoca = epoca +1
print('\033[1;34m')
print(50*"~")
print("pesos finais = ", w)
print("espocas de trinamento = ", epoca)
print(50*"~")
```
